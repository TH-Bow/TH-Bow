# Data Collection with Python Library Pandas

##Import Library pandas

```python
import pandas as pd
```

##Download .csv files to colab


 [audible_data.csv](https://github.com/TH-Bow/audibleBook_Revenue/blob/main/audible_data.csv) AND [audible_transaction.csv](https://www.dropbox.com/scl/fi/f2fqb04ca9amhfwmwjsjt/audible_transaction.csv.zip?rlkey=0z0exq3ho5qy3nytddh6t1et4&dl=0)





```python
#read file "audible_data"
audible_data = pd.read_csv("/content/audible_data.csv")
audible_transaction = pd.read_csv("/content/audible_transaction.csv")
```

##Set Column `Book_ID` to be the index

```python
audible_data = audible_data.set_index("Book_ID")
```

##Delete column Unnamed

```python
audible_data.info()
```

```python
audible_data = audible_data.drop("Unnamed: 0", axis=1)
```

#Transaction Dataframe

##Join table: audible_transaction & audible_data

Use column `Book_ID` to merge transaction dataframe and audible dataframe to see the book's price of each transaction   

```python
transaction = audible_transaction.merge(
    audible_data, how="left", left_on="book_id", right_on="Book_ID"
    )
```

```python
transaction = transaction.drop("Unnamed: 0", axis=1)
```

```python
transaction[:3]
```

#Conversion Rate Dataframe

##Get data from REST API

URL : [Currency conversion API](https://r2de2-workshop-vmftiryt6q-ts.a.run.app/usd_thb_conversion_rate)

```python
import requests
url = "https://r2de2-workshop-vmftiryt6q-ts.a.run.app/usd_thb_conversion_rate"
r = requests.get(url)
result_conversion_rate = r.json()
```

```python
result_conversion_rate
```

The data type of the conversion rate API is dictionay.

it should to be converted to Dataframe before joining to `transaction dataframe`

```python
print(type(result_conversion_rate))
assert isinstance(result_conversion_rate, dict)
```

##Convert to Pandas Dataframe

```python
conversion_rate = pd.DataFrame(result_conversion_rate)
conversion_rate
```

##Rename index column to `Date` column

```python
conversion_rate = conversion_rate.reset_index().rename(columns={"index": "Date"})
conversion_rate
```

```python
conversion_rate.info()
```

##convert `Date` Column to datetime

```python
conversion_rate['Date'] = pd.to_datetime(conversion_rate['Date']).dt.date
```

#Output Dataframe

##Join table: transaction & conversion_rate


first of all copy the `timestamp` column to the `date` column then convert to datetime.

```python
transaction['date'] = transaction['timestamp']
transaction[:3]
```

```python
transaction['date'] = pd.to_datetime(transaction['date']).dt.date
```

Use column `Date` to merge transaction dataframe and conversion rate dataframe

```python
final_df = transaction.merge(conversion_rate, how="left", left_on="date", right_on="Date")
final_df.head()
```

##Delete dollar sign ($)

The data type of the `Price` column is string. ($20.52) delete dollar sign every single row then converted the price column type to float.

```python
final_df["Price"] = final_df.apply(lambda x: x["Price"].replace("$",""), axis=1)
final_df.head()
```

##Coverted to float

converted the `Price` column to float data type

```python
final_df["Price"] = final_df["Price"].astype(float)
final_df.dtypes
```

##Book's price in Thai Baht

Added the `THBPrice` column to the final_df dataframe

```python
final_df["THBprice"] = final_df["Price"] * final_df["conversion_rate"]
final_df.head()
```

delete the `date` column.

```python
final_df = final_df.drop("date", axis=1)
```
